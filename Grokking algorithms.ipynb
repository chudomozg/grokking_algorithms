```python
#Бинарый поиск
def binary_search(list, item):
    low = 0
    high = len(list) - 1

    while low <= high:
        mid = round((low + high) / 2)
        quess = list[mid]

        if quess == item:
            return mid
        if quess > item:
            high = mid - 1
        else:
            low = mid + 1
    return None
```

```python
#Сортировка выбором
def find_smallest(arr):
    smallest_index = 0
    for i in range(1, len(arr)):
        if arr[i] < arr[smallest_index]:
            smallest_index = i
    return smallest_index

def selection_sort(arr):
    sorted_arr = []
    for i in range(len(arr)):
        smallest_index = find_smallest(arr)
        sorted_arr.append(arr.pop(smallest_index))
    return sorted_arr
```

```python
#Рекурсивная сумма 
def recursion_sum(*args):
    args=list(args)
    if len(args) == 1:
        return args[0]
    return args.pop() + recursion_sum(*args)
```

```python
recursion_sum(1,2,3,4,5)
```

```python
#Рекурсивный посчет элементов в листе
def recursion_list_len(lst, count=0):
    if len(lst) == 1: return count+1
    lst.pop()
    return recursion_list_len(lst, count + 1)
```

```python
recursion_list_len([1,2,3,4,5,6,7])
```

```python
#Быстрая сортировка
def quick_sort(arr):
    if len(arr) < 2:
        return arr
    pivot = arr[0]
    less = [i for i in arr[1:] if i <= pivot]
    more = [i for i in arr[1:] if i > pivot]
    return quick_sort(less) + [pivot] + quick_sort(more)
```

```python
quick_sort([1, 34, 7654, 12, 45, 21, 0, 10, 10])
```

```python
#Поиск в ширину
from collections import deque

#ф-ция которая должна сказать является ли человек продавцом манго,
#предположим, что все, чьи имена начинаются на А, умеют проадвать манго
def is_preson_mango_seller(person):
    if person[0] == 'a':
        return True
    return False

def bfs(graph):
    search_queue = deque()
    search_queue += graph['start']
    searched = []
    while search_queue:
        person = search_queue.popleft()
        if is_preson_mango_seller(person):
            print(person + " is seller of mango")
            return True
        searched.append(person)
        search_queue +=graph[person]

    return False #если не нашли
```

```python
import math
#Алгоритм Дейкстры
class Dijkstra_search():
    def __init__(self, graph):
        self.graph = graph
        self.processed = []
        self.costs = {key:value for key, value in graph['start'].items()}
        self.costs['finish'] = math.inf

        self.parents = {key: self.get_start_parents(key) for key in self.costs.keys()}

    def get_start_parents(self, key):
        return None if self.costs[key] == math.inf else "start" 

    def find_lowest_cost_node(self):
        lowest_cost = math.inf
        lowest_cost_node = None
        for node in self.costs:
            cost = self.costs[node]
            if cost < lowest_cost and node not in self.processed:
                lowest_cost = cost
                lowest_cost_node = node
        return lowest_cost_node

    def find_lowest_way(self):
        node = self.find_lowest_cost_node()
        while node is not None:
            cost = self.costs[node]
            neighbors = self.graph[node]
            for n in neighbors.keys():
                new_cost = cost + neighbors[n]
                if new_cost < self.costs[n]:
                    self.costs[n] = new_cost
                    self.parents[n] = node
            self.processed.append(node)
            node = self.find_lowest_cost_node()
        return self.visual_way()

    def visual_way(self):
        out_arr = ['start']
        reverse_parent = {value: key for key, value in self.parents.items()}
        while out_arr[-1] != 'finish':
            out_arr.append(reverse_parent[out_arr[-1]])
        return " -> ".join(out_arr) 
```

```python
gr = {
    "start":{
        "a": 6,
        "b": 2
    },
    "a":{
        "finish": 1
    },
    "b": {
        "a": 3,
        "finish": 5
    },
    "finish": {}
}

ds =  Dijkstra_search(gr)
ds.find_lowest_way()
```

```python
#Жадный алгоритм 
#Задача о покрытии радиостанций штатов

def greedy_alg(states_need, stations):
    final_stations = set()
    while states_need:
        best_station = None
        states_covered = set()
        for station, states in stations.items():
            covered = states_need & states
            if len(covered) > len(states_covered):
                best_station = station 
                states_covered = covered
        
        states_need -= states_covered
        final_stations.add(best_station)
    print(final_stations)
```

```python
states_needed = set(["mt", "wa", "or", "id", "nv", "ut", "ca", "az"])
stations = {
    'kone' : set(['id', "nv", "ut"]),
    'ktwo' : set(['wa', 'id', 'mt']),
    'kthree' : set(['or', 'nv', 'ca']),
    'kfour' : set(['nv', 'ut']),
    'kfive' : set(['ca', 'az'])
}

greedy_alg(states_needed, stations)
```
